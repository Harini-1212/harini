```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# digits.target is used for classification algorithm in ml

digits.target
```

```python
 # a list of names for each feature or attribute in the dataset
 digits.feature_names
```

```python
#dir() returns a list of names in the current local scope.
dir(digits)
```

```python
df=pd.DataFrame(digits.data,columns=digits.feature_names)
df.head()
```

```python
df['target']=digits.target
df.head(30)
```

```python
from sklearn.model_selection import train_test_split
X_test,X_train,y_test,y_train=train_test_split(df.drop('target',axis='columns'),df.target,test_size=0.3)
```

```python
# using RBF kernal.It is a non-linear kernel that allows these algorithms to work effectively in
#higher-dimensionalfeature spaces without explicitly calculating the transformations
```

```python
from sklearn.svm import SVC
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
svm_classifier = SVC(kernel='rbf')
```

```python
svm_classifier.fit(X_train,y_train)
```

```python
svm_classifier.score(X_test,y_test)
```

```python
#liner model.The linear kernel is a simple and commonly used kernel that computes the dot
#product between data points in the original feature space.
```

```python
linear_model = SVC(kernel='linear')
linear_model.fit(X_train,y_train)
```

```python
linear_model.fit(X_train,y_train)
```

```python
linear_model.score(X_test,y_test)
```
